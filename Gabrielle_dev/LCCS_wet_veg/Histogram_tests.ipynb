# Calculate histogram

- Copy NetCDF to location with ArcMap
- Load NetCDF in ArcMap. Right click on properties, select NetCDF, and display the desired percentile
- In arcmap, use Clip (Data Managment) to clip the raster to the Queensland polygons
- Upload clipped raster (now a tif) to jupyter notebook

### Import modules

```python
import xarray as xr
import matplotlib.pyplot as plt
```

### Open clipped image

```python
image = xr.open_rasterio('clipped/clip_9_-29.tif')
image2 = image.squeeze()

### Mask out no data values
image3 = image2.where(image2 != -999)
array = image3.values
```

```python
### Display
plt.imshow(array, vmin = -2000, vmax = 0)
plt.colorbar()
```

### Plot the histogram

```python
fig = plt.figure(figsize=([8,6]))
plt.hist(array,bins=10)
# plt.hist(array, bins = 99, color='k');
# plt.axvline(EventThreshold, color='r', linestyle='--', label=f'Event threshold: {EventThreshold}%');
# plt.ylabel('Count');
# plt.xlabel('Percentage of wetland identified as wet');
# plt.title(f'Histogram of the percentage of {AnalysisLocation} identified as wet across all time');
# plt.legend();
# plt.savefig(WetEventsHistogram, bbox_inches='tight');
```
